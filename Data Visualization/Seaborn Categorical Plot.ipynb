```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Here x is categorical column and y is numerical column and below barplot shows the avegrage total bill amount for category male/female
#barplot is just like a pandas group by and mean aggregate function
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
#Here x is categorical column and y is numerical column and below barplot shows the standard deviation of  total_bill amount for category male/female
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#Count plot shows the number of count for each category just like pandas dataframe count function
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```
