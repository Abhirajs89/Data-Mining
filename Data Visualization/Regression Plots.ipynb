```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='smoker',row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=10)
```
